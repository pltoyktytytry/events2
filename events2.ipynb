```python
date_list = []
name_list = []
location_list = []
information_list = []
follow_list = []
```

```python
for c in event:
    date_list.append(c.find('td',{'class':'col-12 text-dark'}).find('div').text)
    name_list.append(c.find('td',{'class':'col-12 text-break show-related'}).find('h2').text)
    location_list.append(c.find('td',{'class':'col-12 mb-2'}).find('div').text)
    information_list.append(c.find('td',{'class':'col-12 mt-3'}).find('div').text)
    follow_list.append(int(c.find('div',{'class':'small text-center rounded-3 bg-light'}).find('a').text))
```

```python
date_list
```

```python
table = pd.DataFrame([date_list,name_list,location_list,information_list,follow_list]).transpose()
```

```python
table.columns = ['date','name','location','information','follow']
```

```python
table
```

```python
table.to_excel('events.xlsx',engine='openpyxl')
```
